```python
import pandas as pd
```

```python
df = pd.read_csv('ball_by_ball_data.csv')
df2 = pd.read_csv('teams_data.csv')
```

```python
lookup_dict = df2.set_index('team_id')['team_name'].to_dict()
```

```python
replace_columns = ['team_batting','team_bowling']
```

```python
df[replace_columns] = df[replace_columns].replace(lookup_dict)
```

```python
df.drop(
    ['batsman_type','bowler_type','fielders_involved',
     'is_wicket','is_wide_ball','is_no_ball','is_leg_bye','is_bye','is_penalty']
    ,axis = 1,inplace = True)
```

```python
df.to_csv('ball_by_ball_cleaned.csv',index = False)
```
