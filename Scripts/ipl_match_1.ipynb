```python
import pandas as pd
```

```python
df = pd.read_csv('ipl_matches_data.csv')
df2 = pd.read_csv('teams_data.csv')
```

```python
team_id_map = (df2.set_index('team_id')['team_name'].to_dict())
```

```python
team_id_map
```

```python
list_collumns_replace = ['toss_winner','team1','team2','match_winner']
```

```python
df[list_collumns_replace] = df[list_collumns_replace].replace(team_id_map)
```

```python
df.loc[df['win_by_runs'].notna(), 'result'] = (
    'win by ' + df['win_by_runs'].astype('Int64').astype(str) + ' runs'
)

df.loc[df['win_by_wickets'].notna(), 'result'] = (
    'win by ' + df['win_by_wickets'].astype('Int64').astype(str) + ' wickets'
)
```

```python
df.drop(['win_by_runs','win_by_wickets'],axis =1,inplace = True)
```

```python
df.drop(['gender','match_type','season_id','format','event_name','team_type','balls_per_over','overs'],axis = 1,inplace = True)
```

```python
playoffs = df[df['match_number'].isna()]
```

```python
final_idx = playoffs.groupby('season')['match_date'].idxmax()
```

```python
df.loc[final_idx, 'match_number'] = 'Final'
```

```python
df.to_csv('ipl_matches_cleaned.csv',index = False)
```
